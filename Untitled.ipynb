```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import ceil
import seaborn as sns
```

```python
df = pd.read_csv('Train.csv')
df.head().T
```

```python
df.info()
```

```python
build_tech_mean = df['build_tech'][lambda x: ~x.isna()].mean()
metro_dist_mean = df['metro_dist'][lambda x: ~x.isna()].mean()
g_lift_mean = df['g_lift'][lambda x: ~x.isna()].mean()
print(build_tech_mean)
print(metro_dist_mean)
print(g_lift_mean)
```

```python
df['build_tech'].fillna(build_tech_mean, inplace=True)
df['metro_dist'].fillna(metro_dist_mean, inplace=True)
df['g_lift'].fillna(g_lift_mean, inplace=True)
df['date'] = df['date'].apply(lambda x: str.split(x, '-')).apply(lambda x: int(x[0])+(int(x[1])/12))
```

```python
fig, axes = plt.subplots(ceil(df.shape[1]/2)-1, 2, figsize=[16, 52])
for i in range(1, df.shape[1]):
    sns.histplot(df[df.columns[i]], ax = axes[(i-1)//2, (i-1)%2])
    
plt.show()
```

```python
catfeat_with_floor = df.columns.drop('area').drop('id').drop('date').drop('price')
catfeat = catfeat_with_floor.drop('floor').drop('metro_dist').drop('n_photos')
for n in catfeat_with_floor:
    print(n)
    print(pd.value_counts(df[n]))
    ax = plt.subplots(figsize=(7, 2.5))
    plt.xticks(rotation='vertical')
    ax=sns.violinplot(x=n, y="price", data=df, linewidth=1)
    plt.show()
```

```python
#Now I want to turn my categorical variables into dummy variables before I do any regressions
df_with_dummies = df
for category in catfeat:
    series = df[category]
    dummies = pd.get_dummies(series, prefix=category)
    df_with_dummies = pd.concat([df_with_dummies, dummies], axis=1)
    df_with_dummies = df_with_dummies.drop(category, axis=1)
print(df_with_dummies.columns)
df_with_dummies.head()
```

```python
df_with_dummies = df_with_dummies[df_with_dummies['price'] != 1000000]
sns.displot(df_with_dummies['price'])
```

```python
X = df_with_dummies.drop(columns=['price', 'id'])
y = df_with_dummies['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
y_train_log = np.log(y_train)
y_test_log = np.log(y_test)
y_log = np.log(y)
```

```python
class RandomLine():
    
    def __init__(self, n_estimators=20, random_state=42):
        self.n_estimators = n_estimators
        self.lines_ = [] # 
        self.features_idx = []
        self.random_state = random_state

    def fit(self, X, y):
        self.X = X
        self.y = y
        for t in tqdm_notebook(range(self.n_estimators)):               
            #    - 
            LR = LinearRegression()
            #  
            index = np.random.choice(np.arange(X.shape[0]), X.shape[0])
            #  
            features = np.random.choice(np.arange(X.shape[1]), X.shape[1] // 2, replace=False)
            # 
            LR.fit(X[index][:, features], y[index])
            #    
            self.lines_.append(LR)
            #  
            self.features_idx.append(features)
                   
        return self
    
    def predict(self, X):
        answer = np.zeros((len(self.lines_), X.shape[0]), dtype=int)
        #   
        for t in range(len(self.lines_)):
            answer[t] = np.exp(self.lines_[t].predict(X[:, self.features_idx[t]]))
            
        return np.array([np.mean(answer[:, i]) for i in range(answer.shape[1])])
```

### Boosting (not good)

```python
from tqdm import tqdm_notebook

class GradientBoosting():
    
    def __init__(self, n_estimators=100, learning_rate=0.1, random_state=42):  
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.initialization = lambda y: np.mean(y) * np.ones(len(y))
        self.loss_by_iter = [] #     
        self.lines_ = []
        self.random_state = random_state
        
    def mse_grad(self, y, p):
        #  
        return 2. * (p - y) / len(y)

    
    def fit(self, X, y):
        self.X = X
        self.y = y
        b = self.initialization(y)
        prediction = b.copy()
        for t in tqdm_notebook(range(self.n_estimators)):               
            #  - 
            resid = -self.mse_grad(y, prediction)
            #   
            LR = Ridge()
            #    
            LR.fit(X, resid)
            #       
            b = LR.predict(X)
            #   
            self.lines_.append(LR)
            #    (lr * b)
            prediction += self.learning_rate * b
            #     ()
            self.loss_by_iter.append(mean_absolute_error(y, prediction))
            
        return self
    
    def predict(self, X):
        #        
        #         
        pred = np.ones(X.shape[0]) * np.mean(self.y)
        #    * lr
        for t in range(self.n_estimators):
            pred += self.learning_rate * self.lines_[t].predict(X)
            
        return pred
```
